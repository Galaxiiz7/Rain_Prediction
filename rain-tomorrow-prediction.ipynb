```python
def replace_outliers_median(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    df.loc[(df[column] < Q1 - 1.5*IQR) | \
           (df[column] > Q3 + 1.5*IQR), column]\
            = df[column].median()

    return df
```

```python
df = replace_outliers_median(df, 'Evaporation')
df = replace_outliers_median(df, 'WindSpeed9am')
df = replace_outliers_median(df, 'WindSpeed3pm')
df = replace_outliers_median(df, 'WindGustSpeed')
```

```python
Q1 = raining['Rainfall'].quantile(0.25)
Q3 = raining['Rainfall'].quantile(0.75)
IQR = Q3 - Q1
top = Q3 + 1.5*IQR
bot = Q1 - 1.5*IQR
print(bot,top)

df = df.loc[(df['Rainfall'] > bot) &  (df['Rainfall'] < top)]
```

```python
df.fillna(df.select_dtypes('number').median(), inplace=True)
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
plt.figure(figsize = (16,5))

plt.subplot(1, 3, 1)
plt.title('Wind Gust Direction Counts')
sns.countplot(data=df, y='WindGustDir')

plt.subplot(1, 3, 2)
plt.title('Wind 9am Direction Counts')
sns.countplot(data=df, y='WindDir9am')

plt.subplot(1, 3, 3)
plt.title('Wind 3pm Direction Counts')
sns.countplot(data=df, y='WindDir3pm')
```

```python
df[['WindGustDir','WindDir9am','WindDir3pm']] = df[['WindGustDir','WindDir9am','WindDir3pm']].fillna(method = 'ffill')
```

```python
df.info()
```

```python
df.loc[df['Rainfall'] > 1, 'RainToday'] = 'Yes'
df.loc[df['Rainfall'] <= 1, 'RainToday'] = 'No'
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['RainTomorrow'] = df['RainToday'].shift(-1)
df = df[df['Date'].diff().dt.days == 1]
```

```python
df = df.dropna(subset='RainTomorrow')
```

# Data preparation

```python
df_num = df.select_dtypes('number')
df_object = df.select_dtypes('object')
df_object1 = pd.get_dummies(df_object, drop_first=True)
df_prep = pd.concat([df_num, df_object1], axis=1)
```

# Sample data

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df_prep, train_size=0.7)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
x_train = df_train.drop(columns = 'RainTomorrow_Yes')
y_train = df_train['RainTomorrow_Yes']
```

```python
x_test = df_test.drop(columns = 'RainTomorrow_Yes')
y_test = df_test['RainTomorrow_Yes']
```

# Train random forest classifier model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rfmodel = RandomForestClassifier(n_estimators=1000, class_weight='balanced', n_jobs=-1, min_samples_leaf=5, bootstrap=True, min_samples_split= 5, max_samples=0.5, max_depth=30)
rfmodel.fit(x_train,y_train)
```

```python
rfmodel.feature_importances_
```

```python
pd.DataFrame(dict(Feature=df_train.columns[:-1],
                  Value=rfmodel.feature_importances_))\
  .sort_values(by='Value', ascending=False).head(10)
```

# Prediction

```python
y_pred = rfmodel.predict(x_test)
y_pred
```

# Classification report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
